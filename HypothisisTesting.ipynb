```python
recency_p=sample[sample['Response']==1]['Recency'] # conditional statements 
recency_n=sample[sample['Response']==0]['Recency']
```

```python
from scipy.stats import ttest_ind
obj = ttest_ind(recency_p,recency_n)
print(obj.pvalue)
```

```python
#avoid Ho and excepy H1
```

```python
#ANOVA TEST
kidhome_0=sample[sample['Kidhome']==0]['NumWebPurchases']
kidhome_1=sample[sample['Kidhome']==1]['NumWebPurchases']
kidhome_2=sample[sample['Kidhome']==2]['NumWebPurchases']
```

```python
from scipy.stats import f_oneway
obj = f_oneway(kidhome_0,kidhome_1,kidhome_2)
print(obj.pvalue)
```

```python
#avoid HO AND except H1
```

```python
###Chi-Squared test
#Chi-Squared test - is for testing the relationship between 
###two categorical variables. The underlying principal is that
# two categorical variables are independent. 
combine=pd.crosstab(sample['Education'],sample['Response'])
```

```python
combine
```

```python
from scipy.stats import chi2_contingency
result=chi2_contingency(combine)
print(result[1])
```
